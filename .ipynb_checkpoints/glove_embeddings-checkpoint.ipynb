## GloVe Word Embedings ##
This notebook goes through the basics of word embeddings and the famous pre-trained GloVe word embeddings. Also, to create embedding matrix.

### What are word embeddings and why do we use them? ###
Let's assume that the vocabulary size of text corpus(total number of unique words is L). Generally, each word in this set is one hot encoded i.e each word is represented by a vector of L-dimension with all values to be zeros except when the dimension is the same word itself. You will have tens of thousands of unique words in your text vocabulary. Computations with such one-hot encoded vectors will be very inefficient because most values in your one-hot vector will be 0. So, the matrix calculation that will happen in between a one-hot vector and a first hidden layer will result in a output that will have mostly zeros.



Word Embeddings are specifically used to tackle this issue. Instead of representing each word to be a sparse vector of L dimensions, we represent each word in a reduced feature space(for ex: 100 or 300 dimensions). These reduced dimensions help in reduction of the computation and also retain the meaning of similarity i.e. similar words like 'good' , 'great' or 'table', 'desk' have vectors that are very similar to each other.

### How do we obtain them? ###
Word2Vec is one of the most famous and widely used methods to get word embeddings.
It has two approaches:
1. Common Bag Of Words (CBOW):
In this method, we take all the words in context of a target word for which we are trying to calculate the embedding. All these context words are input into a neural network and the network is trained in such a way that the probability of prediction of target word is maximized. Then, we take the output of an intermediate N-dimensional layer from the network for all the context words and average the output to get N-dim word embedding for target word.



2. Skip Gram:
This is some what an opposite apporach to CBOW. In this approach, we take the target word and train a fully connected NN such that the probability of prediction of all the words in context to the target word is maximized.



Both approaches have their own advantages and disadvantages. Skip Gram works well with small amount of data and is found to represent rare words well.
On the other hand, CBOW is faster and has better representations for more frequent words.

### Pretrained Word Embeddings ###
Many of the language tasks perform better if our word embedding model is trained on a larger corpus of text. However this requires lot of compute power and special algorithms to ingest such huge amounts of data. It is generally wiser to go for pre trained word embeddings that are readily available unless the text for your task contains lot of domain specific terms or/and if many words mean different in the context of your domain.

Two of the most popular and widely used pre trained word embeddings are __GloVe__ and __fastText__. 

__GloVe__(Global Vectors for Word Representation) are pre trained word embeddings trained on Wikipedia corpus of 6 billion tokens and vocab size of 400k. There are several embedding dimension options: 50, 100, 200, 300.

For this task of image captioning, we're going to use the GloVe word embeddings which can be downloaded from: http://nlp.stanford.edu/data/glove.6B.zip. 

### Embedding Matrix ###
In keras, we can define an embedding layer with weights of this layer being set to something called an embedding matrix. This matrix has dimensions: vocab size x embedding dim, where vocab size is the number of unique words in the text corpus and embedding dim is the dimension of word embedding in which each word is represented.

Below is a small scale representation of an example embedding matrix, one hot vector and word embedding:



Similar to the image above, we get word embedding for each word(i.e. each row in the embedding matrix) from the pre trained GloVe word embeddings.

In our embedding matrix, each row will represent the word from our corpus and embedding vector for that word is taken from GloVe embeddings.

```python
import pickle
import numpy as np
from scipy import spatial
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from helpers import Helpers
from data_load import DataLoad
from sb_preprocess import ImageDataPreprocess, TextDataPreprocess
```

```python
# opening the 100 dimension word embeddings file
with open('glove_embeddings/glove.6B.200d.txt', 'r') as txt_file:
    glove_embeddings = txt_file.readlines()
```

```python
# making a dictionary with key as word and value as the emedding for that word.
embeddings_dict = {}
for line in glove_embeddings:
    line_content = line.split()
    word, embedding = line_content[0], line_content[1:]
    embeddings_dict[word] = np.asarray(embedding, 'float32')
```

__An embedding for 'embedding'!! Does that sound confusing?__

```python
embeddings_dict['embedding']
```

### Word Similarity ###
The property of word embedding vectors being similar for words having similar meaning/context can be verified with the help of a smple function. The similarity between embedding vecots is measured bu euclidean disance.

```python
def find_closest_words(embedding):
    return sorted(embeddings_dict.keys(), key=lambda each_word: spatial.distance.euclidean(embeddings_dict[each_word], embedding))
```

We are ready to checkout if the emebeddings encode the notion of wrd similarity into them.

Below code calls the function defined above and prints 5 most similar words to the word of interest.

```python
print(find_closest_words(embeddings_dict['laptop'])[1:6])
```

```python
print(find_closest_words(embeddings_dict['president'])[1:6])
```

```python
print(find_closest_words(embeddings_dict['banana'])[1:6])
```

### Visualizing the Vectors ###
To visualize the vectors, we are first going to be using a method known as t-SNE (t-distributed stochastic neighbor embedding), where 't' stands for t-distribution in statitics. t-SNE will allow us to reduce the 100 dimensions of the embedding vector to 2 dimensions

```python
# instantiating a t-SNE object
tsne = TSNE(n_components=2, random_state=0)
```

```python
# list of words we want to visualize
words =  list(embeddings_dict.keys())
# list of embeddings for those words 
vectors = [embeddings_dict[word] for word in words]
#fitting and transforming 100-D data to 2-D
# if remove slicing at the end of vecots, tsne will have to work with all the embeddings of 400k words 
# and this would require enormous compute power.

# visualizing random 500 words
Y = tsne.fit_transform(vectors[113000:113500])
# scatter plot
plt.figure(figsize=(20, 10))
plt.scatter(Y[:, 0], Y[:, 1])
#annotate each point scatter plot with the word
for word, x, y in zip(words, Y[:, 0], Y[:, 1]):
    plt.annotate(word, xy=(x, y), xytext=(0, 0), textcoords="offset points")
plt.grid()
plt.title('Word embeddings visualization')
plt.show()
```

We can see that words with similar meanings/context are clustered together.

### Constructing an Embedding Matrix ###
Let's create and save an Embedding matrix that can later be used as weights to embedding layer.

```python
# first let us create and save a keras tokenizer
# getting a dictionary with image_id as key and list of descriptions as value
descriptions_dict = TextDataPreprocess.get_descriptions_dict(desc_path='Flicker8k_text/Flickr8k.token.txt', clean=True)
with open('pickle_files/all_descriptions_dict.pkl', 'wb') as pickle_file:
    pickle.dump(descriptions_dict, pickle_file)
# saving all the descriptions as text file 'descriptions.txt'
TextDataPreprocess.save_descriptions(descriptions_dict, path_to_save='descriptions.txt')
```

```python
# creating a dictionary of training descriptions to fit the tokenizer
train_descriptions_dict = DataLoad.load_descriptions(path_to_ids='Flicker8k_text/Flickr_8k.trainImages.txt', descriptions_path='descriptions.txt')
with open('pickle_files/train_descriptions_dict.pkl', 'wb') as pickle_file:
    pickle.dump(train_descriptions_dict, pickle_file)
```

```python
# loading saving the tokenizer
word_counts, word_to_id, id_to_word = Helpers.get_mappings(train_descriptions_dict, freq_threshold = 10)
with open('pickle_files/word_to_id.pkl', 'wb') as pickle_file:
    pickle.dump(word_to_id, pickle_file)
with open('pickle_files/id_to_word.pkl', 'wb') as pickle_file:
    pickle.dump(id_to_word, pickle_file)
vocab_size = len(word_to_id) + 1
print('size of training vocabulary: {}'.format(vocab_size))
```


Reading the 100d word embeddings file and creating a dictionary such that key is the word and value is the embedding of that word.

```python
embedding_dim = 200
with open('glove_embeddings/glove.6B.200d.txt', 'r') as txt_file:
    glove_embeddings = txt_file.readlines()
embeddings_dict = {}
for line in glove_embeddings:
    line_content = line.split()
    word, embedding = line_content[0], line_content[1:]
    embeddings_dict[word] = np.asarray(embedding, 'float32')
```

Creating an embedding matrix such that each row represents the embedding of a particluar word in the vocabulary of our training data.

```python
embedding_matrix = np.zeros((vocab_size, embedding_dim))
# looping through each unique word in the trainig set.
for word, i in word_to_id.items():
    if word == 'startseq':
        embedding_matrix[i] = np.array([0.1]*embedding_dim)
    elif word == 'endseq':
        embedding_matrix[i] = np.array([0.2]*embedding_dim)
    else:
        embedding_vector = embeddings_dict.get(word)
        if embedding_vector is not None:
            # words not found in embedding index will be all-zeros.
            embedding_matrix[i] = embedding_vector
```

```python
# saving the embedding matrix in the 'npy' format
np.save('embedding_matrix.npy', embedding_matrix)
```
